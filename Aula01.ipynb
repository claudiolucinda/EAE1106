### Como ccomputadores armazenam nmeros

Para entender a diferena entre computao de preciso simples e dupla,  importante entender o papel da preciso na cincia da computao. Imagine realizar um clculo usando um nmero irracional (como pi) e incluindo apenas dois dgitos  direita do ponto decimal (3.14). Voc obteria um resultado mais preciso se fizesse o clculo incluindo dez dgitos  direita do ponto decimal (3,1415926535).



Para computadores, esse nvel de preciso  chamado de preciso e  medido em dgitos binrios (bits) em vez de casas decimais. Quanto mais bits usados, maior a preciso. A representao de grandes nmeros em binrios de computador requer um padro para garantir que no haja grandes discrepncias nos clculos. Assim, o Instituto de Engenheiros Eltricos e Eletrnicos (IEEE) desenvolveu o Padro IEEE para Aritmtica de Ponto Flutuante (IEEE 754).

Em geral, um nmero ser representado de forma aproximada com um nmero fixo de nmeros significantes e escalonado usando um expoente em uma base (usualmente dois, ou 10 ou 16). Esses nmeros possuem a seguinte forma:

$ S \times B^e$

Em que $S$  o significando, $B$  um nmero inteiro maior ou igual a dois, e $e$ um expoente, um nmero inteiro. Por exemplo:
$1.2345=1234 \times 10^{-4}$. O termo Ponto Flutuante se refere ao fato que o ponto decimal (tambm chamado de ponto binrio) pode flutuar; ou seja, pode ser colocado em qualquer lugar em relao aos dgitos significantes do nmero. Esta posio  indicada pelo expoente, e a representao em ponto flutuante de um nmero pode ser pensada como uma forma de notao cientfica. 

A preciso nessa representao de ponto flutuante  dada pela quantidade de bits dedicado ao armazenamento do significando e do expoente -- quanto maior o significando, mais precisa  a representao (menor  o espao entre dois nmeros), e quanto maior o expoente, mais alto  o maior valor que pode ser representado.


Existem trs componentes do IEEE 754:

* O sinal - 0 representa um nmero positivo; 1 representa um nmero negativo.
* O expoente - O expoente  usado para representar expoentes positivos e negativos. 
* O significando - Tambm conhecida como mantissa, a mantissa representa os bits de preciso do nmero.

Usando esses componentes, o IEEE 754 representa nmeros de ponto flutuante de duas maneiras: formato de preciso simples e formato de preciso dupla. Embora ainda existam vrias maneiras de representar nmeros de ponto flutuante, o IEEE 754  o mais comum porque geralmente  a representao mais eficiente de valores numricos.

##### O que  o formato de ponto flutuante de preciso simples?
O formato de ponto flutuante de preciso simples usa 32 bits (um bit pro sinal, 8 pro expoente e 24 pro significando) de memria do computador e pode representar uma ampla gama de valores numricos. Muitas vezes referido como FP32, esse formato  melhor usado para clculos que no sofrem um pouco de aproximao.

##### O que  o formato de ponto flutuante de preciso dupla?
O formato de ponto flutuante de preciso dupla, por outro lado, ocupa 64 bits (um bit pro sinal, 11 pro expoente e 52 pro significando) da memria do computador e  muito mais preciso do que o formato de preciso simples. Esse formato  frequentemente chamado de FP64 e usado para representar valores que exigem um intervalo maior ou um clculo mais preciso.

Embora a preciso dupla permita mais preciso, ela tambm requer mais recursos computacionais, armazenamento de memria e transferncia de dados. O custo de usar esse formato nem sempre faz sentido para todos os clculos.

## IDE 

IDE  um acrnimo para Integrated Development Environment, em portugus, Ambiente de Desenvolvimento Integrado. Ele  um programa que rene ferramentas necessrias para a construo de outros softwares.

A utilizao de um IDE ajuda muito os programadores e empresas, pois torna mais rpido o desenvolvimento de aplicaes, aumentando a produtividade e reduzindo custos.

Existem IDE's especficas para plataformas, e outras que so mais flexveis. As mais famosas so o Sublime Text e o Visual Studio Code. 

Quais so os componentes tpicos de uma IDE:

* Editor de cdigo-fonte: Permite edio do cdigo nas linguagens de programao suportadas pelo IDE. 

* Preenchimento inteligente: Esse  um recurso trazido pelos IDEs que agiliza o desenvolvimento, pois escreve automaticamente trechos do cdigo como, por exemplo, comandos de funo.

* Compilador: O compilador que voc escreveu em uma determinada linguagem de programao para a linguagem de mquina, de modo que os computadores o entendam.

* Debugger:  outra ferramenta que contribui para no cdigo-fonte, melhorando o desempenho do programa.

* Gerao de cdigo: Com esse recurso,  possvel predefinir trechos de cdigos para serem usados de modelo em outros projetos, agilizando o desenvolvimento de trabalhos futuros.


Como podemos ver, os IDEs renem diversas ferramentas que tornam mais simples a vida dos programadores. Vamos listar alguns benefcios que a utilizao desses programas pode trazer para seus projetos:

* Reduz o tempo gasto em cada aplicao;
* Permite o desenvolvimento de um cdigo mais limpo, organizado e legvel;
* Aumenta a produtividade dos desenvolvedores e das empresas;
* Reduz a quantidade de bugs e falhas no cdigo-fonte;
* Rene diversas ferramentas em um s lugar.
